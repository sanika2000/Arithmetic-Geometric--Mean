# Arithmetic-Geometric Mean
Arithmetic-Geometric Mean(AGM) is a number where aithmetic and geometric mean converges. Arithmetic Mean is always greater than or equal to geometric mean. So at some point arithmetic and geometric mean will converge.

**Arithmetic-Geometric mean of 2 nos**

```python
import math
exp = 1e-10;
a = 24;
b = 6;
an = (a+b)/2
gn = math.sqrt(a*b);
while abs(an-gn)>exp and an-gn!=0:
    c = an;
    an = (an + gn)/2.0;
    gn = math.sqrt(c * gn);
print("AGM of 24 and 6 is " + str(an));
```

**AGM over range of numbers**

```python
import math
from matplotlib import pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
agm = [];
x_values = [];
y_values = [];
exp = 1e-10;
k = 0;
print("give lower bound of range");
r1 = int(input());
print("give upper bound of range");
r2 = int(input());
r3 = r2 + 1;
for i in range(r1,r3):
    j = r3-i-1;
    x_values.extend([i for k in range(j)]);
#print(x_values);
for i in range(r1,r3):
    for j in range(i+1,r3):
      y_values.append(j);  
#print(y_values);


for i in range(r1,r3):
    for j in range(i+1,r3):
        an = (i+j)/2;
        gn = math.sqrt(i * j);
        while abs(an-gn)>exp and an-gn!=0:
            b = an;
            an = (an + gn)/2.0;
            gn = math.sqrt(b * gn);
       
        agm.append(an);



fig = plt.figure();

ax = Axes3D(fig);
ax.set_title('Plot for Arithmetic-Geometric mean of all pairs in range');
ax.set_xlabel("first number");
ax.set_ylabel("second number");
ax.set_zlabel("AGM");
ax.scatter(x_values,y_values,agm);
```
